```python
import pandas as pd
import numpy as np
import os
```

```python
data_joint_model = pd.read_csv('../results/joint-modeling/data_joint_modeling_all.csv')
```

```python
df_sorted = data_joint_model.sort_values('rt').reset_index(drop=True)
```

```python
df_sorted['bin_id'] = np.arange(len(df_sorted)) // 20
```

```python
bin_means = df_sorted.groupby('bin_id')[['ams', 'pams', 'slps']].mean()
```

```python
data_joint_model_rt_bin = df_sorted.merge(bin_means, left_on='bin_id', right_index=True, suffixes=('', '_bin_rt'))
```

```python
PATH_RESULTS_JOINT = os.path.join('..', 'results', 'joint-modeling')
data_joint_model_rt_bin.to_csv(
        os.path.join(PATH_RESULTS_JOINT, 'data_joint_modeling_all_rt_bin.csv'),
        index=False
    )
```
